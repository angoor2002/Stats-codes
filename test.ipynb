```python
import pandas as pd
import numpy as np

# Assuming 'A' and 'B' are the columns in your DataFrame
df = pd.read_excel(r"HDR21-22_Statistical_Annex_HDI_Trends_Table.xlsx")

# Create a contingency table
categories_A = df.iloc[:, 2].unique()
categories_B = df.iloc[: ,4].unique()

contingency_table = pd.DataFrame(index=categories_A, columns=categories_B)


for cat_A in categories_A:
    for cat_B in categories_B:
        count = np.sum((df.iloc[:, 0] == cat_A) & (df.iloc[:, 1] == cat_B))
        contingency_table.loc[cat_A, cat_B] = count

        
print(contingency_table)
```
